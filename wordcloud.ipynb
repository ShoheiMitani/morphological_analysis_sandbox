```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
from unicodedata import normalize

def plot(text):
    # 環境に合わせてフォントのパスを指定する。
    fpath = " ~/Library/Fonts/ヒラギノ角ゴシック W3.ttc"

    wordcloud = WordCloud(font_path=fpath, width=900, height=500).generate(text)

    plt.figure(figsize=(15,12))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()

df = pd.read_csv('similarity_grouping_result.csv').sort_values('total_count', ascending=False)

# 抽出したくない文字たちのフィルタリング調整
ng_list = ["ケース", "No Key", "アン", "ホルダー", "ちゃん", "ちゃ", "ポーチ", "ぬいぐるみ", "フィギュア", "ポーチ", "ガチャ", "フィギュア", "ストラップ",
                 "カチューシャ", "バッグ", "一番くじ", "グッズ", "値下げ", "バッジ", "ポストカード","バッジ", "スター", "ペン", "トート", "アクリル",  "ポケット",
                 "カード", "日本未発売", "チェーン", "フレンズ", "シール"]
selected_df = df[~df['frequent_word'].isin(ng_list)].nlargest(50, 'total_count')

total_num = selected_df['total_count'].sum()

text_list = []
for index, row in selected_df.iterrows():
    rate = row['total_count'] / total_num * 100
    if rate > 4:
        rate = 4 #=> 気にくわないけど数を抑えないと崩れるバグがあるっぽい

    for x in range(int(rate)):
       text_list.extend([normalize('NFKC', row['frequent_word'])])

text = u" ".join(text_list)
plot(text)

selected_df.reset_index(drop=True)
```
